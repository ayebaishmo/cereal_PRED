both used to create pipelines in scikit-learn, but they have some differences:

1. Pipeline()


More explicit — You provide names for each step.


Allows you to name each stage (e.g., 'imputer', 'scaler', 'model')


Useful when you need to refer to steps by name later (e.g., using pipeline.named_steps['scaler'])

2. make_pipeline()


Simpler syntax — You don’t name the steps manually.


scikit-learn automatically assigns names based on the class name (e.g., 'standardscaler', 'logisticregression').


pipeline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),

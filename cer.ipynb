```python
"""2. Scale the numeric features
        a. Instantiate a StandardScaler for the numeric features."""

scaler = StandardScaler()

```

```python
"""b. Fit the scaler on X_train_num_imputed"""
# first identify numerical columns and impute them
df.dtypes
```

```python
# numerical_columns = df.select_dtypes(include = ["float", "int"]).columns """wrong formart it might not capture int64"""

num_cols =  X_train.select_dtypes(include=["number"]).columns
num_cols
```

```python
# Lets check out the null values in numeric columns of trianing set before we impute
null_counter_train = X_train[num_cols].isna().sum()
print("Columns with null values in trainning set \n", null_counter_train[null_counter_train>0])
```

```python
# Impute the columns 
# instatiate the imputer
impute_median = SimpleImputer(strategy='median')
```

```python
# fit the imputer 
impute_median.fit(X_train[num_cols])
```

```python
# view the median values calculated during the .fit step
impute_median.statistics_
```

```python
# confirm that these are median values
X_train[num_cols].median().values
```

```python
# Change settings to ouput a dataframe
from sklearn import set_config
set_config(transform_output='pandas')
```

```python
# Lets transform to fill the null values for X_train set since we still habe the null values in the columns 
X_train_num_imputed = impute_median.transform(X_train[num_cols])
```

```python
#check for missing values again to confirm that we have filled all of them with median
null_counter_train = X_train_num_imputed[num_cols].isna().sum()
print("Columns with null values in trainning set \n", null_counter_train[null_counter_train>0])
```

```python
X_train_num_imputed
```

```python
"""b. Fit the scaler on X_train_num_imputed"""
scaler.fit(X_train_num_imputed)
# we can the view the mean and SD
print("Mean",scaler.mean_)
print("std",scaler.scale_)
```

```python
"""Transform the numeric training data (X_train_num_imputed) and save the output 
as "X_train_num_scaled"""

X_train_num_scaled = scaler.transform(X_train_num_imputed)
X_train_num_scaled
```

```python
"""Transform the numeric test data (X_test_num_imputed) and save the 
transformed data as "X_test_num_scaled"""

# start by imputing and substuting the null values before we apply scaler 
# Lets check for null values in the test set before we impute it 
null_counter_test = X_test[num_cols].isna().sum()
print("Columns with null values in testing set \n", null_counter_test[null_counter_test>0])
```

```python
# lets now transform it to fill all null values in the calories column 
X_test_num_imputed = impute_median.transform(X_test[num_cols])
```

```python
# check if we have substitied all the null values in the calories column 
null_counter_test = X_test_num_imputed[num_cols].isna().sum()
print("Columns with null values in testing set \n", null_counter_test[null_counter_test>0])
```

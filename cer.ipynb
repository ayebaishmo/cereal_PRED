Standard scaling, also known as standardization, is a technique used to normalize the features of a dataset so that they have a mean of 0 and a standard deviation of 1

This is done by subtracting the mean of each feature and then dividing by the standard deviation.

learning 


Normalization is a data preprocessing technique used to scale the values of features into a specific range, typically [0, 1]. It ensures that no feature dominates the others due to scale differences, which helps many machine learning algorithms perform better.

```python
"""2. Scale the numeric features
        a. Instantiate a StandardScaler for the numeric features."""

scaler = StandardScaler()

```

```python
"""Scale the ordinal features
Instantiate another StandardScaler for the ordinal features."""
scaler_ordinal_feature = StandardScaler()
```

```python
# The categorical columns 
cat_cols = ['mfr', 'type', 'shelf']
```

```python
# Lets instatiate a impute method
impute_cat_cols = SimpleImputer(strategy='constant', fill_value='NA')
```

```python
# Lets fit the columns now 
impute_cat_cols.fit(X_train[cat_cols])
```

```python
# Now lets transform the categorical columns
X_train_cat_imputed =  impute_cat_cols.transform(X_train[cat_cols])
# before they were imputed 
print(X_train['mfr'].unique())
print(X_train['type'].unique())
print(X_train['shelf'].unique())
```
